```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

data = pd.read_csv('Position_Salaries.csv')

X = data.iloc[:, 1:2].values
y = data.iloc[:, 2].values

linear_regressor = LinearRegression()
linear_regressor.fit(X, y)

poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
poly_regressor = LinearRegression()
poly_regressor.fit(X_poly, y)

plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.scatter(X, y, color='blue')
plt.plot(X, linear_regressor.predict(X), color='red')
plt.title('Simple Linear Regression: Position vs Salary')
plt.xlabel('Position Level')
plt.ylabel('Salary')

plt.subplot(1, 2, 2)
plt.scatter(X, y, color='blue')
plt.plot(X, poly_regressor.predict(poly_reg.fit_transform(X)), color='green')
plt.title('Polynomial Regression: Position vs Salary')
plt.xlabel('Position Level')
plt.ylabel('Salary')

plt.tight_layout()
plt.show()
```
